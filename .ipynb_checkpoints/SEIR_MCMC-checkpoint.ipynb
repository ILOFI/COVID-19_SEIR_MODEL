```python
def get_ci(predict_value):
    s = np.random.poisson(predict_value, 1000000)
    return np.percentile(s, [2.5, 97.5])
```

```python
def SEIR_MODEL(infect_data, init_S, init_E = 1, init_I = 1):
    T = len(infect_data)
    case_data = infect_data
    init_E = case_data[3]

    beta = Uniform('beta', 1e-9, 0.1, value = 0.0005)
    mu = Uniform('mu', 1e-9, 0.1, value = 0.0001)
    beta2 = Uniform('beta2', 1e-9, 0.1, value = 0.0001)
    mu2 = Uniform('mu2', 1e-9, 0.1, value = 0.0001)
    gamma = Uniform('gamma', 1e-9, 0.1, value = 0.0001)
    s0 = Uninformative('s0', value = init_S)
    e0 = Uniform('e0', 0, (init_E +1 )* 5, value = init_E)
    i0 = Uniform('i0', 0, (init_I +1) * 5, value = init_I)

    
    @deterministic
    def sim(beta = beta, mu = mu, gamma = gamma, s0 = s0, e0 = e0, i0 = i0, beta2 = beta2, mu2 = mu2):
        S = np.zeros(T)
        E = np.zeros(T)
        I = np.zeros(T)
        S[0] = s0
        E[0] = e0
        I[0] = i0
        cumulative_cases = np.zeros(T)
        cumulative_cases[0] = infect_data[0]

        for j in range(1, T):
            if j > 11:
                S[j] = S[j - 1] - beta2 * S[j - 1] * I[j - 1] 
                E[j] = E[j - 1] + beta2 * S[j - 1] * I[j - 1]  - mu2 * E[j - 1]
                I[j] = I[j - 1] + mu2 * E[j - 1] - gamma * I[j - 1]
                cumulative_cases[j] = cumulative_cases[j - 1] + mu2 * E[j - 1]
            else:
                S[j] = S[j - 1] - beta * S[j - 1] * I[j - 1]
                E[j] = E[j - 1] + beta * S[j - 1] * I[j - 1] - mu * E[j - 1]
                I[j] = I[j - 1] + mu * E[j - 1] - gamma * I[j - 1]
                cumulative_cases[j] = cumulative_cases[j - 1] + mu * E[j - 1]

        return cumulative_cases
    cases = Lambda('cases', lambda sim = sim : sim)
    A = Poisson('A', mu = cases, value = case_data, observed = True)
    return locals()
```

```python
select_city = origin_data['city'].unique()
city_params = []
for cityname in select_city:
    
    startdate = '2020/3/18' # 24
    print(cityname)
    cityid = city_ID[cityname]
    init_S = population[cityid, 2]
    greater = np.sum(epidemic[cityid, 2:] > 0)
    window = max(7, greater)
    window = min(window, len(alldates) - alldates.index(startdate)) #wuhan 23
#     window = min(8, window)
    infect_data = epidemic[cityid, -window:]
    
    print('Infect:', infect_data)
    
    mod = SEIR_MODEL(infect_data[:30], init_S)
    mc = MCMC(mod)
    mc.use_step_method(AdaptiveMetropolis, [mod['beta'], mod['mu'], mod['gamma'], mod['s0'], mod['e0'], mod['i0'], mod['beta2'], mod['mu2']])
    mc.sample(iter = 55000, burn = 5000, thin = 500, verbose = 0) # old 55000
    
    plt.figure(figsize=(8, 6))
    plt.title('SEIR Model for {}'.format(cityname))
    plt.plot(mod['case_data'], 's', mec = 'black', color = 'black')
    plt.plot(mod['cases'].stats()['mean'], color = 'blue', linewidth = 2)
    plt.plot(mod['cases'].stats()['95% HPD interval'][0], color = 'blue', linewidth = 1, linestyle = 'dotted')
    plt.plot(mod['cases'].stats()['95% HPD interval'][1], color = 'blue', linewidth = 1, linestyle = 'dotted')
    plt.ylabel('Cumulative Number of Cases')
    plt.xlabel('Time (days)')
    plt.grid(True)
    plt.savefig('./figures/{}.png'.format(cityname), dpi=400)
    plt.show()
    
    param_dict = {}
    for param_name in ['beta', 'mu', 'gamma', 's0', 'e0', 'i0', 'beta2', 'mu2']:
        param_dict[param_name] = mod[param_name].stats()['mean']
    city_params.append(param_dict)
#     break
```
